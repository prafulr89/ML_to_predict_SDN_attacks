```python
import pandas as pd

data = pd.read_csv("Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv")

data["Label"] = data["Label"].astype('category')

data["Label"] = data["Label"].cat.codes

print (data.shape)
print (data.columns)
#print (data.head)
print (data.info())
#print (data.describe())
```

```python
#Manual Drop & Encoding 

data = data.drop("Flow ID", axis=1)
data = data.drop("Source IP", axis=1)
data = data.drop("Destination IP", axis=1)

data["Timestamp"] = data["Timestamp"].astype('category')

data["Timestamp"] = data["Timestamp"].cat.codes

print (data.shape)
print (data.columns)
#print (data.head)
print (data.info())
#print (data.describe())
```

```python
import numpy as np
import matplotlib.pyplot as plt
data_1 = data.isnull().sum()
print (data_1)
data_2 = data_1[data_1>(0.05*len(data))]
print (data_2)
print(data_2.index)
data = data.drop(data_2.index, axis=1)
data = data.fillna(0);
data.to_csv("pp.csv", index=False)
print (data.shape)
print (data.columns)
```

```python
class_label = 'Label'
data_3 = data.select_dtypes(include=['integer','float']).corr()[class_label].abs()
print (data_3)
x = np.arange(data_3.shape[0])
y = data_3
plt.plot(x, y, color ="green")
plt.savefig('corr.png')
plt.show()
data_3.to_csv("cv.csv")
data = data.drop(data_3[data_3<0.1].index, axis=1)
print (data.shape)
print (data.columns)
print (data.head)
print (data.info())
print (data.describe())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from math import sqrt
```

```python
from sklearn.preprocessing import StandardScaler
X = data[list(data.columns)[:-1]].select_dtypes(include=['integer','float'])
print(np.any(np.isnan(X)))    #value should be FALSE
print(np.all(np.isfinite(X))) #value should be TRUE
scaler = StandardScaler()
scaled = scaler.fit_transform(X)
print(scaled)
```

```python
#RF
from sklearn.ensemble import RandomForestClassifier
X = data[list(data.columns)[:-1]]
print(np.any(np.isnan(X)))
print(np.all(np.isfinite(X)))
print(X.shape)
y = data['Label']
x_train, x_test, y_train, y_test = train_test_split(X,y)
clf = RandomForestClassifier(n_estimators=100, max_depth=20)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
#print(y_pred)
cm = confusion_matrix(y_test, y_pred)
acc = clf.score(x_test, y_test)
rmse = sqrt(mean_squared_error(y_test, y_pred))
print("RF")
print ("Acc : ", acc)
print (cm)
summary = classification_report(y_test, y_pred)
plot_confusion_matrix(clf, x_test, y_test)
print(summary)
k = 8
cv_scr = cross_val_score(clf, X, y, cv=k,scoring='r2')
print("Cross Validation Score(k={})".format(k))
print(cv_scr)
print("Cross Validation Mean: {}".format(np.mean(cv_scr)))
plt.show()
probs = clf.predict_proba(x_test)
preds = probs[:,1]
fpr, tpr, th = metrics.roc_curve(y_test, preds)
roc_auc = metrics.auc(fpr, tpr)
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.plot(fpr, tpr, 'g', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'center')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
#NB
from sklearn.naive_bayes import GaussianNB
X = data[list(data.columns)[:-1]].select_dtypes(include=['integer','float'])
y = data['Label']
x_train, x_test, y_train, y_test = train_test_split(X,y)
print(y_test)
clf = GaussianNB()
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
acc = clf.score(x_test, y_test)
rmse = sqrt(mean_squared_error(y_test, y_pred))
print("NB")
print ("Accuracy : ", acc)
print ("Confusion Matrix")
print (cm)
summary = classification_report(y_test, y_pred)
plot_confusion_matrix(clf, x_test, y_test)
print(summary)
plt.show()
probs = clf.predict_proba(x_test)
preds = probs[:,1]
fpr, tpr, th = metrics.roc_curve(y_test, preds)
roc_auc = metrics.auc(fpr, tpr)
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.plot(fpr, tpr, 'g', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'center')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
X = data[list(data.columns)[:-1]]
y = data['Label']
x_train, x_test, y_train, y_test = train_test_split(X,y)
import keras
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(x_train.shape[1],)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10, batch_size=1)
test_loss, test_acc = model.evaluate(x_test, y_test)
y_pred = np.array(model.predict(x_test))
y_pred =np.argmax(y_pred,axis=1)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Test accuracy:', test_acc)
summary = classification_report(y_test, y_pred)
print(summary)
```
